### Data Cleaning (Applying the changes needed according to the data assessment above)
- Remove duplicates  
- Drop unnecessary columns (id, imdb_id, cast, homepage, director, tagline, keywords, overview, budget_adj, revenue_adj)
- Fill/Remove missing values (genres, revenue, budget, production companies)
- Profit column to be calculated (revenue - budget) and inserted

```python
#Remove duplicates
df.drop_duplicates(inplace=True)
df.shape
```

```python
#Drop unnecessary columns
df.drop(columns=['id', 'imdb_id', 'cast', 'homepage', 'director', 'tagline',
                 'keywords', 'overview', 'budget_adj', 'revenue_adj'], inplace = True)
```

```python
df.shape
```

```python
#Fill/Remove missing values
#Drop rows where 'production_companies' is null
df['production_companies'].isnull().sum()
```

```python
df.dropna(axis=0, subset=['production_companies'], inplace=True)
df.shape
```

```python
#Drop rows where 'genres' is null
df['genres'].isnull().sum()
```

```python
df.dropna(axis=0, subset=['genres'], inplace=True)
df.shape
```

```python
#Replace "0" in 'revenue' and 'budget' with NaN
df['budget'].replace(0, np.nan, inplace=True)
df['revenue'].replace(0, np.nan, inplace=True)
```

```python
df.shape
```

```python
#remove NaNs from 'revenue' and 'budget' (I can't keep them because they are affecting my study and I can't replace
#them with real numbers because movies budgets/revenues don't depend on each other so they should be dropped)
df.dropna(subset=['budget', 'revenue'], inplace=True)
df.shape
```

```python
df.info()
```

```python
#change datatype of 'budget' & 'revenue' back to int instead of float
df['revenue'] = df['revenue'].astype(int)
df['budget'] = df['budget'].astype(int)
```

```python
#Create profit column
df['profit'] = df['revenue'] - df['budget']
df.shape
```

```python
df.info()
```

```python
#Explore data in hist
df.hist(figsize=(15,8));
```

```python
def genreMax(column):
    return df[df[column] == max(df[column])][["genres", 'original_title', column]]

def genreMin(column):
    return df[df[column] == min(df[column])][["genres", 'original_title', column]]
```

```python
genreMin('vote_average')
```

##### From above we can conclude that the least voted Genres combination goes to "Animation, Action, Comedy and Family"

### Research Question 4: Which year has the most/least profitable movie?

```python
#Group release_year to be able to answer the question above
year_series_max = df.groupby(['release_year']).profit.max()
year_series_max
```

```python
year_series_max.max()
```

```python
year_series_max[year_series_max == 1910483648]
```

##### From above we can conclude that the year 2009 witnessed the most profittable movie

```python
year_series_min = df.groupby(['release_year']).profit.min()
year_series_min
```

```python
year_series_min.min()
```

```python
year_series_min[year_series_min == -413912431]
```

##### From above we can conclude that the year 2010 witnessed the least profittable movie

```python
#Plot for the top 5 earning years (from the series: year_series_max)
year_series_max = year_series_max.sort_values(ascending=False)
year_series_max_top5 = year_series_max[:5]
year_series_max_top5
```

```python
fig = plt.figure(figsize=(15,8))
year_series_max_top5.plot(kind='bar', rot='0')
plt.title("Top 5 Profitable Release Years", fontsize=20)
plt.ylabel("Profit (In billions)")
plt.xlabel("Release Years");
```

##### This clearly shows that the year 2009 is standing out in the top 5 profitable years while 2011 is the least of them

### Research Question 5: What is the relationship between runtime and positive profit earnings?

We can show this using a scatter plot but first let us take all the movies with positive profit (above 0)

```python
positive_profit = df[df['profit'] > 0]
positive_profit.head(3)
```

```python
x = positive_profit.runtime
y = positive_profit.profit
```

```python
fig = plt.figure(figsize=(15, 10), tight_layout=True, dpi=200)
plt.title("Relation between runtime and profitablitiy", fontsize=40)
plt.ylabel("Profit (In billions)", fontsize=25)
plt.xlabel("Runtime (in minutes)", fontsize=25)
plt.scatter(x,y);
```

##### This clearly shows that the relationship between the movie runtime and profit is somehow directly propotional
